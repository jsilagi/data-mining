```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = [[2,2], [1,3], [2,3], [2,5], [3,5], [3,6], [5,1], [5,2], [5,6], [6,1], [6,2]]
```

```python
distortions = []

for i in range(1,12):
    model = KMeans(n_clusters=i)
    model.fit(data)
    distortions.append(model.inertia_)
```

```python
plt.plot(range(1,12), distortions, 'o-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('Finding the ideal k')
plt.show()
```
