# Load the data

```python
# Load in dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler

df = pd.read_excel("C:/Users/silag/OneDrive/Desktop/DataMining/Concrete_Data.xls")
```

# Functions for univariate models

```python
# Univariate predict

def predict(x, m, b):
    y = (m*x) + b
    return y
```

```python
# Mean-squared error function

def mse(x, y, m, b):
    s = np.sum( (y-((m*x)+b)) * (y-((m*x)+b)) )
    l = s/len(x)
    return l
```

```python
# Partial derivative of MSE loss function w.r.t. parameter b

def partialB(x, y, predictedy):
    s = np.sum(-2*(y - predictedy))
    l = s/len(x)
    return l
```

```python
# Partial derivative of MSE loss function w.r.t. parameter m

def partialM(x, y, predictedy):
    s = np.sum((-2*x) * (y - predictedy)) 
    l = s/len(x)
    return l
```

# Gradient descent and regression for univariate models

```python
# Randomly sample 900 training observations and 130 testing observations
train = df.sample(n=900, random_state=63130)
test = df.drop(train.index)

# Iterate over the 8 factors
for col in df.drop(columns = 'Concrete compressive strength(MPa, megapascals) '):
    
    # Split train and test into x and y
    x_train = train[col]
    y_train = train['Concrete compressive strength(MPa, megapascals) ']
    
    x_test = test[col]
    y_test = test['Concrete compressive strength(MPa, megapascals) ']
    
    
    # Learning rate
    alpham = 0.001
    alphab = 1

    # Initialize m and b
    m = 0
    b = 0

    # Perform gradient descent
    for i in range(10000):      
        
        # Keep m and b within reasonable bounds
        if b > 100 or b < -100:
            alphab = alphab * 0.1
            b = 0
        if m > 100 or m < -100:
            alpham = alpham * 0.1
            m = 0
        
        # Update m and b
        m -= alpham * partialM(x_train, y_train, predict(x_train, m, b))
        b -=  alphab * partialB(x_train, y_train, predict(x_train, m, b))

    # Report results of univariate models
    print(col)
    print('b ' + str(b))
    print('m ' + str(m))
    print('mse ' + str(mse(x_test, y_test, m, b)))
    print('training r2 ' + str(r2_score(y_train, predict(x_train, m, b))))
    print('testing r2 ' + str(r2_score(y_test, predict(x_test, m, b))))
    plt.scatter(x_train, y_train)
    plt.plot(x_train, predict(x_train, m, b), color='red')
    plt.xlabel(col)
    plt.ylabel('Concrete Compressive Strength (MPa)')
    plt.show()
```

# Functions for multivariate model

```python
# Multivariate predict

def multiPredict(x, m, b):
    y = np.dot(m.transpose(),x.transpose()) + b
    return y
```

```python
# Multivariate MSE function

def multiMse(x, y, m, b):
    s = np.sum( (y-(np.dot(m.transpose(),x.transpose())+b)) * (y-(np.dot(m.transpose(),x.transpose())+b)) )
    l = s/len(x)
    return l
```

```python
# Multivariate partial of loss function w.r.t b

def multiPartialB(x, y, predictedy):
    s = np.sum(-2*(y - predictedy))
    l = s/len(y)
    return l
```

```python
# Multivariate partial of loss function w.r.t m

def multiPartialM(x, y, predictedy):
    arr = []
    for col in x.transpose():
        s = np.sum((-2*col) * (y - predictedy))
        l = s/len(y)
        arr.append(l)  
    return(np.array(arr))
```

# Gradient descent and regression for multivariate model

```python
# Randomly sample 900 training observations and 130 testing observations
train = df.sample(n=900, random_state=63130)
test = df.drop(train.index)

# Split train and test into x and y
x_train = train.loc[:, train.columns!='Concrete compressive strength(MPa, megapascals) ']
y_train = train['Concrete compressive strength(MPa, megapascals) ']

x_test = test.loc[:, test.columns!='Concrete compressive strength(MPa, megapascals) ']
y_test = test['Concrete compressive strength(MPa, megapascals) ']

# Learning rate
alpham = 0.0000001
alphab = 0.1

# Initialize m and b
m = np.zeros(8)
b = 0

# Perform gradient descent
for i in range(10000):

    # Update m and b
    m -= alpham * multiPartialM(x_train.values, y_train, multiPredict(x_train.values, m, b))
    b -=  alphab * multiPartialB(x_train.values, y_train, multiPredict(x_train.values, m, b))

# Report result for multivariate model
print('Multivariate Model')
print('b ' + str(b))
print('m ' + str(m))
print('mse ' + str(multiMse(x_test, y_test, m, b)))
print('training r2 ' + str(r2_score(y_train, multiPredict(x_train, m, b))))
print('testing r2 ' + str(r2_score(y_test, multiPredict(x_test, m, b))))
```
